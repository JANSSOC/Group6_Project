```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
path = 'Project 1 CSV Data/Poverty-Trends_1998-through-2016.csv'

data = pd.read_csv(path)

data
```

```python
data1 = data.drop(['Geography Name', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006',], axis=1)
meck = data1.iloc[61]
meck
wake = data1.iloc[93]
wake
```

```python
pov = data1.iloc[[0, 1, 61, 93]] 
pov
```

```python

pov['2007'] = pov['2007'].str.rstrip('%').astype('float')
pov['2008'] = pov['2008'].str.rstrip('%').astype('float')
pov['2009'] = pov['2009'].str.rstrip('%').astype('float')
pov['2010'] = pov['2010'].str.rstrip('%').astype('float')
pov['2011'] = pov['2011'].str.rstrip('%').astype('float')
pov['2012'] = pov['2012'].str.rstrip('%').astype('float')
pov['2013'] = pov['2013'].str.rstrip('%').astype('float')
pov['2014'] = pov['2014'].str.rstrip('%').astype('float')
pov['2015'] = pov['2015'].str.rstrip('%').astype('float')
pov['2016'] = pov['2016'].str.rstrip('%').astype('float')
```

```python
pov
```

```python
years = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
us = pov.iloc[0]
nc = pov.iloc[1]
meckl = pov.iloc[2]
wak = pov.iloc[3]
plt.figure(figsize=(7,7))
# USA poverty rate
plt.plot(years, us , color="green", label="US Poverty Rate")

# NC poverty rate
plt.plot(years, nc, color="r", label="NC Poverty Rate")
# Mecklenburg poverty rate
plt.plot(years, meckl, color="lightskyblue", label="Mecklenburg Poverty Rate")
# Wake poverty rate
plt.plot(years, wak, color="magenta", label="Wake Poverty Rate")
# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")
plt.gca().yaxis.grid(True)
plt.title("Poverty Rate")
plt.xlabel("Years")
plt.ylabel("Number of Wins/Losses")

# Print our chart to the screen
plt.savefig("Poverty Rate Graph")
plt.show()
```
