```python
import csv
import pandas as pd

path = 'Project 1 CSV Data/2010-2017-NC-County-Population-Estimates_growth_rate.csv'

data = pd.read_csv(path)

data
```

```python
#droping the usless column on the far left and far right
data1 = data.drop('Unnamed: 0', 1)
data2 = data1.drop('Unnamed: 9', 1)
#droped the blank top rows with NaN values
data3 = data2.dropna(0,'all')
data3
```

```python

#Renaming the Columns to make them more clear 
data4 = data3.rename(index=str, columns={'Unnamed: 1':'Name',
                                        'Unnamed: 2': 'April 1,2010 Total Population Est.',
                                        'Unnamed: 3': 'July 1, 2017 Total Poluation Est.',
                                        'Unnamed: 4': 'Numeric Change 10-17',
                                        'Unnamed: 5': 'Growth Rate 10-17',
                                        'Unnamed: 6': 'County Ranking: Pop',
                                        'Unnamed: 7': 'Numeric Change Rank',
                                        'Unnamed: 8': 'Growath Rate Rank'})

#Dropping the column names found within the data frame and also resetting the index
#Side Note: I am creating copies of the data frames so that if the group/ me wants to they can access what the df looked like before the changes
data5 = data4.reset_index(drop=True)
data6 = data5.drop([0,1])
data7 = data6.reset_index(drop=True)
data7
```

```python
#Summary of Meck County
Meck_data = data7.loc[60,:]
Meck_data
```

```python
#In df form
Meckdf = data7.loc[[60]]
Meckdf
```

```python
#From here on sort by ranking if we want to compare counties if not ignore
#I need to drop the North Carolina numbers as to not have a row with NAN as it messes with sort by
data8 = data7.dropna(0,'any')
#I have to reset the column type to integers before we can sort by
#This comes up with a warning but its okay it does not mess with the data it is just not ideal
data8['County Ranking: Pop'] = data8['County Ranking: Pop'].astype('int')
#I sort by the population ranking
pop_rank = data8.sort_values(by='County Ranking: Pop')
pop_rank.head()
```

```python
ndata = data7.dropna(0,'any')
#I have to reset the column type to integers before we can sort by
#This comes up with a warning but its okay it does not mess with the data it is just not ideal
ndata['Numeric Change Rank'] = ndata['Numeric Change Rank'].astype('int')
#I sort by the population ranking
num_rank = ndata.sort_values(by='Numeric Change Rank')
num_rank.head()
```

```python
gdata = data7.dropna(0,'any')
#I have to reset the column type to integers before we can sort by
#This comes up with a warning but its okay it does not mess with the data it is just not ideal
gdata['Growath Rate Rank'] = gdata['Growath Rate Rank'].astype('int')
#I sort by the population ranking
growth_rank = gdata.sort_values(by='Growath Rate Rank')
growth_rank.head()
```
