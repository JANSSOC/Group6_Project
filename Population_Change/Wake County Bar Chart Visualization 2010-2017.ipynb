```python
%matplotlib notebook

import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
wake_pop = [901059, 906884, 929845, 952982, 974938, 998914, 1023859, 1049143, 1072203]
wake_pop
x_axis = np.arange(len(wake_pop))
```

```python
from matplotlib.pyplot import figure
figure(num=1, figsize=(8, 5), dpi=160, facecolor='w', edgecolor='k')
plt.bar(x_axis, wake_pop, color='r', alpha=0.5, align="center")
plt.gca().yaxis.grid(True)
plt.show()
```

```python
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["4-2010", "7-2010", "7-2011", "7-2012", "7-2013", "7-2014", "7-2015", "7-2016", "7-2017"])
```

```python
plt.title("Population Change of Wake County from 2010-2017")
plt.xlabel("Years")
plt.ylabel("Population")
```
