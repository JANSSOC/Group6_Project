```python

%matplotlib notebook

import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis_data = [2010, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
x_axis_data
```

```python
wake_pop = [901059, 906884, 929845, 952982, 974938, 998914, 1023859, 1049143, 1072203]
wake_pop
```

```python
from matplotlib.pyplot import figure
figure(num=1, figsize=(8, 5), dpi=160, facecolor='w', edgecolor='k')
plt.plot(x_axis_data, wake_pop)
plt.gca().yaxis.grid(True)
plt.show()
plt.savefig('Plot Images/compare_line.png')
```

```python
meck_pop = [919650, 923202, 944943, 968204, 991619, 1011315, 1034442, 1057237, 1076837]
meck_pop
```

```python
plt.plot(x_axis_data, wake_pop, color="magenta")
plt.plot(x_axis_data, meck_pop, color="lightskyblue")
plt.legend(["Mecklenberg County", "Wake County"])
plt.show()
```

```python
plt.title("Comparison of Mecklenburg and Wake County Population Growth from 2010-2017")
plt.xlabel("Years")
plt.ylabel("Population")
```
