```python
%matplotlib notebook

import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
meck_pop = [919650, 923202, 944943, 968204, 991619, 1011315, 1034442, 1057237, 1076837]
meck_pop
x_axis_data = np.arange(len(meck_pop))
```

```python
from matplotlib.pyplot import figure
figure(num=1, figsize=(8, 5), dpi=160, facecolor='w', edgecolor='k')
plt.bar(x_axis_data, meck_pop, color='b', alpha=0.5, align="center")
plt.gca().yaxis.grid(True)
plt.show()

```

```python
tick_locations = [value for value in x_axis_data]
plt.xticks(tick_locations, ["4-2010", "7-2010", "7-2011", "7-2012", "7-2013", "7-2014", "7-2015", "7-2016", "7-2017"])
```

```python
plt.title("Population Change of Mecklenburg County from 2010-2017")
plt.xlabel("Years")
plt.ylabel("Population")
```
